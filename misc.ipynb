```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
width = np.linspace(0, 10, 100)
length = np.linspace(0, 10, 100)
height = np.linspace(0, 1, 100)
volume = width * length * height

flows = np.repeat(100, 10)
```

```python
plt.plot(width, height)

for i in flows:
    
    volume = volume - i
    plt.plot(i)
```
